# QUESTION 1 

```python
import pandas as pd

url = "https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user"

users = pd.read_csv(url, sep='|')

users.head()

mean_age_per_occupation = users.groupby('occupation')['age'].mean()
print("Mean age per occupation:")
print(mean_age_per_occupation)

def male_ratio(data):
    males = data[data['gender'] == 'M']
    return len(males) / len(data)

male_ratio_per_occupation = users.groupby('occupation').apply(male_ratio).sort_values(ascending=False)
print("\nMale ratio per occupation (sorted):")
print(male_ratio_per_occupation)

min_max_age_per_occupation = users.groupby('occupation')['age'].agg(['min', 'max'])
print("\nMinimum and maximum ages per occupation:")
print(min_max_age_per_occupation)

mean_age_per_occupation_sex = users.groupby(['occupation', 'gender'])['age'].mean()
print("\nMean age per occupation and sex:")
print(mean_age_per_occupation_sex)

def gender_percentage(data):
    gender_counts = data['gender'].value_counts(normalize=True) * 100
    return gender_counts

gender_percentage_per_occupation = users.groupby('occupation').apply(gender_percentage)
print("\nPercentage of women and men per occupation:")
print(gender_percentage_per_occupation)
```

```python
# Import the necessary libraries
import pandas as pd

# Import the dataset
url = "https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/02_Filtering_%26_Sorting/Euro12/Euro_2012_stats_TEAM.csv"


euro12 = pd.read_csv(url)


goals = euro12['Goals']
print("Goals column:")
print(goals)


num_teams = euro12['Team'].nunique()
print(f"\nNumber of teams that participated in Euro2012: {num_teams}")


num_columns = euro12.shape[1]
print(f"\nNumber of columns in the dataset: {num_columns}")


discipline = euro12[['Team', 'Yellow Cards', 'Red Cards']]
print("\nDiscipline DataFrame:")
print(discipline)


sorted_teams = discipline.sort_values(by=['Red Cards', 'Yellow Cards'], ascending=[True, True])
print("\nTeams sorted by Red Cards and then Yellow Cards:")
print(sorted_teams)


mean_yellow_cards = euro12['Yellow Cards'].mean()
print(f"\nMean Yellow Cards per Team: {mean_yellow_cards:.2f}")


teams_more_than_6_goals = euro12[euro12['Goals'] > 6]
print("\nTeams that scored more than 6 goals:")
print(teams_more_than_6_goals)


teams_start_with_g = euro12[euro12['Team'].str.startswith('G')]
print("\nTeams that start with 'G':")
print(teams_start_with_g)


first_7_columns = euro12.iloc[:, :7]
print("\nFirst 7 columns:")
print(first_7_columns)


all_except_last_3 = euro12.iloc[:, :-3]
print("\nAll columns except the last 3:")
print(all_except_last_3)


shooting_accuracy_selected = euro12[euro12['Team'].isin(['England', 'Italy', 'Russia'])][['Team', 'Shooting Accuracy']]
print("\nShooting Accuracy from England, Italy, and Russia:")
print(shooting_accuracy_selected)
```

```python
import pandas as pd
import numpy as np


file_path = r'C:\Users\INNOCENT\Desktop\Test2\New folder\wind.txt'
data = pd.read_csv(file_path, delim_whitespace=True)

data.columns = ['Yr', 'Mo', 'Dy'] + list(data.columns[3:])

data['Yr'] = data['Yr'].apply(lambda x: x + 1900 if x < 100 else x)  # Fix years like 61 -> 1961
data['Date'] = pd.to_datetime(data[['Yr', 'Mo', 'Dy']].astype(str).agg('-'.join, axis=1))
data = data.set_index('Date').drop(columns=['Yr', 'Mo', 'Dy'])

def fix_year(year):
    return year - 100 if year > 2000 else year
data.index = data.index.map(lambda x: x.replace(year=fix_year(x.year)))

print(f"Index data type: {data.index.dtype}")

missing_values = data.isna().sum()
print("\nMissing values for each location:")
print(missing_values)

total_non_missing = data.notna().sum().sum()
print(f"\nTotal non-missing values: {total_non_missing}")

overall_mean = data.mean().mean()
print(f"\nOverall mean windspeed across all locations and times: {overall_mean:.2f}")

loc_stats = data.aggregate(['min', 'max', 'mean', 'std']).T
print("\nLocation statistics (min, max, mean, std):")
print(loc_stats)

day_stats = data.aggregate(['min', 'max', 'mean', 'std'], axis=1)
print("\nDaily statistics (min, max, mean, std):")
print(day_stats)

january_data = data[data.index.month == 1]
january_avg = january_data.mean()
print("\nAverage January windspeed for each location:")
print(january_avg)

yearly_data = data.resample('Y').mean()
print("\nYearly frequency downsampled data:")
print(yearly_data)

monthly_data = data.resample('M').mean()
print("\nMonthly frequency downsampled data:")
print(monthly_data)

weekly_data = data.resample('W').mean()
print("\nWeekly frequency downsampled data:")
print(weekly_data)

weekly_stats = data.resample('W').aggregate(['min', 'max', 'mean', 'std']).iloc[:52]
print("\nWeekly statistics (min, max, mean, std) for the first 52 weeks:")
print(weekly_stats)
```

```python
import pandas as pd

url = "https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv"
chipo = pd.read_csv(url, sep='\t')


print("First 10 entries:")
print(chipo.head(10))

num_observations = chipo.shape[0]
print(f"\nNumber of observations in the dataset: {num_observations}")

num_columns = chipo.shape[1]
print(f"\nNumber of columns in the dataset: {num_columns}")

print("\nColumn names:")
print(chipo.columns)

print("\nDataset index:")
print(chipo.index)

most_ordered_item = chipo.groupby('item_name')['quantity'].sum().idxmax()
print(f"\nMost-ordered item: {most_ordered_item}")

most_ordered_quantity = chipo.groupby('item_name')['quantity'].sum().max()
print(f"\nQuantity of the most-ordered item: {most_ordered_quantity}")

most_ordered_choice_description = chipo.groupby('choice_description')['quantity'].sum().idxmax()
print(f"\nMost-ordered choice description: {most_ordered_choice_description}")

total_items_ordered = chipo['quantity'].sum()
print(f"\nTotal items ordered: {total_items_ordered}")

chipo['item_price'] = chipo['item_price'].str.replace('$', '').astype(float)

print("\nItem price type after conversion:")
print(chipo['item_price'].dtypes)

chipo['revenue'] = chipo['quantity'] * chipo['item_price']
total_revenue = chipo['revenue'].sum()
print(f"\nTotal revenue for the period: ${total_revenue:.2f}")

total_orders = chipo['order_id'].nunique()
print(f"\nTotal orders made: {total_orders}")

average_revenue_per_order = total_revenue / total_orders
print(f"\nAverage revenue per order: ${average_revenue_per_order:.2f}")

different_items_sold = chipo['item_name'].nunique()
print(f"\nNumber of different items sold: {different_items_sold}")
```

```python
import matplotlib.pyplot as plt
import pandas as pd

data = {
    "Year": [1867, 1870, 1880, 1900, 1950, 2000, 2014],
    "Marriages": [4.5, 4.7, 6.0, 8.5, 11.2, 7.8, 6.2],
    "Divorces": [0.3, 0.4, 0.5, 1.2, 4.1, 4.0, 3.2],
}
df = pd.DataFrame(data)

plt.figure(figsize=(10, 6))
plt.plot(df['Year'], df['Marriages'], label='Marriages per capita', color='blue', marker='o')
plt.plot(df['Year'], df['Divorces'], label='Divorces per capita', color='red', marker='o')

plt.xlabel('Year')
plt.ylabel('Number per Capita')
plt.title('Marriages and Divorces per Capita in the U.S. (18672014)')
plt.legend()
plt.grid(True)

# Show the plot
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

data = {
    "Year": [1900, 1950, 2000],
    "Marriages": [8.5, 11.2, 7.8],
    "Divorces": [1.2, 4.1, 4.0],
}
df = pd.DataFrame(data)

plt.figure(figsize=(10, 6))

x_positions = range(len(df['Year']))

plt.bar(x_positions, df['Marriages'], width=0.4, label='Marriages per capita', color='blue', align='center')
plt.bar(
    [x + 0.4 for x in x_positions],
    df['Divorces'],
    width=0.4,
    label='Divorces per capita',
    color='red',
    align='center',
)

plt.xlabel('Year')
plt.ylabel('Number per Capita')
plt.title('Comparison of Marriages and Divorces per Capita (1900, 1950, 2000)')
plt.xticks([x + 0.2 for x in x_positions], df['Year'])  # Adjust tick positions
plt.legend()

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

data = {
    "Actor": ["Arnold Schwarzenegger", "Sylvester Stallone", "Bruce Willis", "Jason Statham", "Keanu Reeves"],
    "Kill Count": [842, 786, 503, 361, 297],
}

df = pd.DataFrame(data)

df = df.sort_values(by="Kill Count", ascending=True)

plt.figure(figsize=(10, 6))
plt.barh(df["Actor"], df["Kill Count"], color="crimson")

plt.xlabel("Kill Count")
plt.ylabel("Actor")
plt.title("Deadliest Actors in Hollywood")
plt.grid(axis="x", linestyle="--", alpha=0.7)

for index, value in enumerate(df["Kill Count"]):
    plt.text(value + 10, index, str(value), va='center', ha='left')

plt.tight_layout()
plt.show()
```
